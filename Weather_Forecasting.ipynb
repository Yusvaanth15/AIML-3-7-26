Remove Outliers

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df[[
        "Temperature",
        "Humidity",
        "Wind_Speed",
        "Pressure"
    ]]
)

plt.title("Before Removing Outliers")

plt.show()
```

Removing Outliers

```python
columns=[
    "Temperature",
    "Humidity",
    "Wind_Speed",
    "Pressure"
]

for col in columns:

    Q1=df[col].quantile(0.25)

    Q3=df[col].quantile(0.75)

    IQR=Q3-Q1

    lower=Q1-1.5*IQR

    upper=Q3+1.5*IQR

    df=df[
        (df[col]>=lower)&
        (df[col]<=upper)
    ]

print("Outliers Removed Successfully")
print(df.shape)
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df[[
        "Temperature",
        "Humidity",
        "Wind_Speed",
        "Pressure"
    ]]
)

plt.title("After Removing Outliers")

plt.show()
```
